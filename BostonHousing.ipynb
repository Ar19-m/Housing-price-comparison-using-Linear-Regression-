Importing dataset and descibing our dataset

```python
import pandas as pd
Housing = pd.read_csv("C:\\Users\\adity\\Desktop\\Python\\BostonHousing.csv")
Housing.describe()
```

Feature names

```python
print(Housing.info())
```

check for missing value

```python
Housing.isna()
```

Creating X and Y matrices

```python
X = Housing.drop(['medv'], axis=1)
print("X matrix is \n", X)
```

```python
Y = Housing.medv
print("Y matrix is \n", Y)
```

Data split

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score, accuracy_score, root_mean_squared_error
```

Splitting data in 85:15 ratio

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size=0.15)
```

training model fitting

```python
model = linear_model.LinearRegression()
model.fit(X_train, Y_train)
```

apply training model on test set

```python
Y_pred = model.predict(X_test)
```

Display model performance

```python
r2_score_linear = r2_score(Y_test,Y_pred)
print("Coefficient:", model.coef_)
print("Inetrcept:", model.intercept_)
print("Mean Squared Error is: ", root_mean_squared_error(Y_test, Y_pred))
print("Coefficient of Determination(R^2): ", r2_score_linear)
```

Make scatter plot using seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.scatterplot(x =Y_test, y= Y_pred)
ax.set(xlabel = "Y_test", ylabel = "Y_predicted", title = "Linear Regression scatter plot")
plt.plot(color = "red", lw = 2)
```

Build Gradient boosting model

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Initialize gradient boosting

```python
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42, learning_rate=0.02, max_depth=3)
```

Train the model

```python
gb_model.fit(X_train, Y_train)
```

Make predictions

```python
Y_pred_gb = gb_model.predict(X_test)
```

Performance Evaluation

```python
# Evaluate the performance
r2_score_grad = r2_score(Y_test, Y_pred_gb)
mean_squared_error_grad = mean_squared_error(Y_test, Y_pred_gb)
print(' Gradient Boosting Performance ')
print(f'Mean squared error (MSE): {mean_squared_error_grad:.2f}')
print(f'Coefficient of determination (R^2): {r2_score_grad:.2f}')
```

```python
# Visualization: Actual vs. Predicted
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Y_test, y=Y_pred_gb, alpha=0.9)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', lw=2)
plt.xlabel('tested Median Housing value')
plt.ylabel('Predicted Median Housing value')
plt.title('Gradient Boosting: Actual vs Predicted')
plt.show()
```

Summary table

```python
# Final Comparison Table
import pandas as pd

results = {
    'Model': ['Linear Regression', 'Gradient Boosting (Tuned + Scaled)'],
    'R2 Score': [r2_score_linear, r2_score_grad],
    'Status': ['Acceptable performance ', 'Best performer'],
}

df_final = pd.DataFrame(results)
print(" Final Model Comparison ")
display(df_final)
```

Linear regression performs competitively due to predominantly linear relationships between housing prices and socio-economic indicators. Boosting models capture interactions but risk overfitting due to limited sample size.

Linear Regression model feature ranking


```python
#Extract feature importance
import pandas as pd
importances = gb_model.feature_importances_
import numpy as np

feature_names = X.columns  

# Create a DataFrame for visualization
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Which markers  matter most?')
plt.show()
```

Gradient Boosting feature ranking

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor

gb_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('gb', GradientBoostingRegressor(
        n_estimators=100,
        learning_rate=0.02,
        max_depth=3,
        random_state=42
    ))
])

# Fit the pipeline
gb_pipeline.fit(X_train, Y_train)

# Extract trained GB model
gb_model = gb_pipeline.named_steps['gb']

# Feature importance
gb_importance_2 = pd.DataFrame({
    'Feature': feature_names,
    'Importance': gb_model.feature_importances_
}).sort_values(by='Importance', ascending=False)


# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=gb_importance_2)
plt.title('Which markers  matter most?')
plt.show()
```

Both linear regression and gradient boosting identify similar top features, indicating that dominant predictors exert consistent influence across linear and non-linear models. This is an indicator of stable underlying structure  artifacts.

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(
    gb_pipeline,
    X_test,
    Y_test,
    n_repeats=10,
    random_state=42,
    scoring='r2'
)

perm_importance = pd.DataFrame({
    'Feature': feature_names,
    'Importance': result.importances_mean
}).sort_values(by='Importance', ascending=False)

perm_importance
```

*Feature Importance Interpretation: *

Gradient Boosting identifies LSTAT and RM as the dominant predictors of housing prices, jointly accounting for the majority of predictive power. The agreement with linear regression indicates stable, global relationships rather than model-specific interactions. Other socio-economic and environmental variables contribute marginally once these dominant factors are accounted for, suggesting redundancy rather than irrelevance.
